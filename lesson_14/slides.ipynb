### Что такое нейронная сеть?

1. Начинаем с перцептрона (простая разделяющая гиперплоскость).
2. Функция активации (шаг, сигма) (добавить изображений)
3. Берем несколько перцептронов (пример с XOR)
4. Организуем иерархически. (пример с лицами, изображение, понимание что извлекаемые участки - это отображение весов нейронов)

1. Многослойный перцептрон не имеет знания о соседстве точек.

### [Классификация в Keras](http://machinelearningmastery.com/tutorial-first-neural-network-python-keras/) на примере [индийских данных о диабете](https://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes).

````python
from keras.models import Sequential, load_model
from keras.layers import Dense
import numpy as np

# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
# load pima indians dataset
dataset = np.loadtxt(
    "pima-indians-diabetes.csv",
    delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:, 0:8]
Y = dataset[:, 8]
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, nb_epoch=10, batch_size=10)
# evaluate the model
scores = model.evaluate(X, Y)
print("\n %s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))
````

````python
model_name = 'indians-diabetes-12-10-6-1.h5'
model.save(model_name)
model2 = load_model(model_name)
predictions = model2.predict(X)
print (predictions[:10])
# round predictions
threshold = 0.5
predictions_formatted = [x[0] > threshold for x in predictions]
print (predictions_formatted[:10])
````

### [А теперь посмотрим как Keras справится с изображениями и пример СNN](https://elitedatascience.com/keras-tutorial-deep-learning-in-python)

````python
import numpy as np
np.random.seed(123)
````

````python
# Sequential model type from Keras. This is simply a linear stack of neural network layers
from keras.models import Sequential
````

````python
# "core" layers from Keras. These are the layers that are used in almost any neural network:
from keras.layers import Dense, Dropout, Activation, Flatten
````

````python
# CNN layers from Keras. These are the convolutional layers that will help us efficiently train on image data
from keras.layers import Convolution2D, MaxPooling2D
````

````python
# utilities
from keras.utils import np_utils
````

````python
from keras.datasets import mnist
 
# Load pre-shuffled MNIST data into train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
````

````python
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(X_train[0])
````

````python
print X_train.shape
````

````python
X_train = X_train.reshape(X_train.shape[0], 1, 28, 28)
X_test = X_test.reshape(X_test.shape[0], 1, 28, 28)
print X_train.shape
````

````python
X_train
````

````python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
````

````python
print y_train.shape
````

````python
print y_train[:10]
````

````python
# Convert 1-dimensional class arrays to 10-dimensional class matrices
Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)
Y_train[:10]
````

````python
print Y_train.shape
````

[Примеры архитектур на Keras](https://github.com/fchollet/keras/tree/master/examples)

````python
model = Sequential()
# input layer
model.add(Convolution2D(32, 3, 3, activation='relu', input_shape=(1,28,28)))
````

````python
model.add(Convolution2D(32, 3, 3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
````

````python
model.output_shape
````

````python
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
````

````python
model.output_shape
````

````python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
````

Различные варианты [функций потерь](https://keras.io/objectives/) и [оптимизаторов](https://keras.io/optimizers/)

````python
model.fit(X_train, Y_train, 
          batch_size=32, nb_epoch=2, verbose=1)
````

Для кроссвалидации всегда можно [совмещать Keras с sklearn](https://github.com/fchollet/keras/issues/1711#issuecomment-185801662)

### [Многослойные перцептроны есть и в Sklearn](http://scikit-learn.org/stable/modules/neural_networks_supervised.html)

````python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_mldata
from sklearn.neural_network import MLPClassifier

mnist = fetch_mldata("MNIST original")
# rescale the data, use the traditional train/test split
X, y = mnist.data / 255., mnist.target
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]

# mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=400, alpha=1e-4,
#                     solver='sgd', verbose=10, tol=1e-4, random_state=1)
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=10, alpha=1e-4,
                    solver='sgd', verbose=10, tol=1e-4, random_state=1,
                    learning_rate_init=.1)

mlp.fit(X_train, y_train)
print("Training set score: %f" % mlp.score(X_train, y_train))
print("Test set score: %f" % mlp.score(X_test, y_test))

fig, axes = plt.subplots(4, 4)
# use global min / max to ensure all weights are shown on the same scale
vmin, vmax = mlp.coefs_[0].min(), mlp.coefs_[0].max()
for coef, ax in zip(mlp.coefs_[0].T, axes.ravel()):
    ax.matshow(coef.reshape(28, 28), cmap=plt.cm.gray, vmin=.5 * vmin,
               vmax=.5 * vmax)
    ax.set_xticks(())
    ax.set_yticks(())
````

Может использоваться для регрессии

````python
from sklearn.neural_network import MLPRegressor 
import math
 
# create Trainig Dataset
train_x=[[x, x**2, x**3, x**4] for x in np.linspace(0,2,100)]
train_y=[math.sin(x[0]) for x in train_x]
 
#create neural net regressor
reg = MLPRegressor(hidden_layer_sizes=(32,),solver="lbfgs", max_iter=1000)
reg.fit(train_x,train_y)
 
#test prediction
test_x=[[x, x**2, x**3, x**4] for x in np.linspace(1,2,10)]
 
predict=reg.predict(test_x)
print ("_Input_\t_output_")
for i in range(len(test_x)):
    print ("  ",test_x[i][0],"---->",predict[i]," actual ", math.sin(test_x[i][0]))
    
plt.plot(np.asarray(test_x)[:,0], predict)
plt.plot(np.asarray(test_x)[:,0], [math.sin(test_x[i][0]) for i in range(len(test_x))])
````
